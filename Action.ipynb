```python
class ActionInterface:

    def __init__(self):
        self.current_action = None

    def execute(self, decision, brain):

        action = decision["action"]

        if action == "PICK":

            obj = decision["object"]
            obj_id = decision.get("id", "unknown")
            position = decision["position"]

            print(f"[ACTION] Attempting PICK")
            print(f"         Object : {obj}")
            print(f"         ID     : {obj_id}")
            print(f"         Position (x,y,z): {position}")

            # Simulate grasp success probability
            success = self.simulate_grasp(position)

            if success:
                print("[ACTION] Pick successful")
                brain.pick_success()
            else:
                print("[ACTION] Pick failed")
                brain.pick_failed()

        elif action == "SEARCH":

            print("[ACTION] Searching environment...")

        elif action == "FOCUS":

            target = decision.get("object", "target")
            print(f"[ACTION] Focusing sensors on {target}")

        elif action == "WAIT":

            print("[ACTION] Waiting for grasp completion")

    # -----------------------------------
    # Simulated grasp success
    # -----------------------------------

    def simulate_grasp(self, position):

        distance = position[2]

        # closer objects easier to grasp
        if distance < 0.8:
            probability = 0.9
        elif distance < 1.2:
            probability = 0.75
        else:
            probability = 0.5

        return np.random.rand() < probability
```
